                                       NETFLIX DATA CLEANING, ANALYSIS AND VISUALIZATION :---------

```python
#IMPORTING THE NECESSARY LIBRARIES WHICH I WILL WORK UPON
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as  plt 
from matplotlib import image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import seaborn as sns 
```

```python
df = pd.read_csv('netflix1.csv')
df                                                         #THE NETFLIX DATA TABLE 
```

```python
df.shape
```

                                                  | DATA CLEANING AND REPROCESSING |

```python
#DATA CLEANING AND REPROCESSING STARTS HERE....!!
df.columns = df.columns.str.title() #To convert the column heads into sentence case 
```

```python
df.head() #To check if the above formatting was successful
```

```python
df['Date_Added'] = pd.to_datetime(df['Date_Added'])  #To change the date-like strings into date-time objects
```

```python
df.head() #To check if the above formatting was successful 
```

```python
df.info() #Basic information about the dataset
```

```python
df.fillna({'Director': 'Unknown', 'Country': 'Unknown'}, inplace = True)
```

                                                          VISUALIZATIONS 

```python
#VISUALIZING THE NETFLIX DATA USING SEABORN AND MATPLOTLIB LIBRARIES.....!!
type_total_streams = df['Type'].value_counts()
type_total_streams
```

```python
#Visualizing the top contents by type: Movies & TV shows 
plt.figure(figsize= (8, 6))
sns.barplot(x=type_total_streams.index, y=type_total_streams.values, color = 'pink')
text_box = plt.title('TOP CONTENTS BY TYPE-- MOVIES AND TV SHOWS', fontdict={'fontsize': 14, 'fontweight': 'bold', 'fontname': 'Trebuchet MS', 'color': 'white'})
text_box.set_bbox(dict(facecolor='red', alpha= 1, edgecolor = 'black'))
logo = mpimg.imread(r'C:\Users\hisar\OneDrive\Desktop\Jupyter\7124274_netflix_logo_icon.png')
plt.figimage(logo, 0, 0, zorder=0, alpha=0)
imagebox = OffsetImage(logo, zoom= 0.3)
ab = AnnotationBbox(imagebox, (0.5, 1.2), frameon=False, xycoords='axes fraction', boxcoords="axes fraction", pad=0)
plt.gca().add_artist(ab)
plt.xlabel('Type')
plt.ylabel('Total number of streams')
plt.show()
```

```python
ratings_on_content = df['Rating'].value_counts()
ratings_on_content
```

```python
#Visualizing the ratings based on content 
plt.figure(figsize= (8, 6))
sns.barplot(x=ratings_on_content.index, y=ratings_on_content.values, color = 'red')
text_box = plt.title('RATINGS BASED ON CONTENT:', fontdict={'fontsize': 14, 'fontweight': 'bold', 'fontname': 'Garamond', 'color': 'black'})
text_box.set_bbox(dict(facecolor='red', alpha= 1, edgecolor = 'black'))
logo = mpimg.imread(r'C:\Users\hisar\OneDrive\Desktop\Jupyter\pngwing.com.png')
plt.figimage(logo, 0, 0, zorder=0, alpha=0)
imagebox = OffsetImage(logo, zoom= 0.07)
ab = AnnotationBbox(imagebox, (0.5, 1.2), frameon=False, xycoords='axes fraction', boxcoords="axes fraction", pad=0)
plt.gca().add_artist(ab)
plt.xlabel('RATINGS', color = 'purple')
plt.ylabel('DISTRIBUTION', color = 'purple')
plt.xticks(rotation = 60)
plt.show()
```

```python
countries = df['Country'].value_counts().head(20)
countries
```

```python
#Visualizing the top 20 countries which has streamed netflix contents
plt.figure(figsize = (10, 8))
sns.barplot(x=countries.values, y=countries.index, hue = countries.values, palette = 'bright')
text_box = plt.title('TOP 20 COUNTRIES WHICH STREAM NETFLIX CONTENTS [2008-2020]', fontdict={'fontsize': 14, 'fontweight': 'bold',  'fontname': 'Times New Roman', 'color': 'black'})
logo = mpimg.imread(r'C:\Users\hisar\OneDrive\Desktop\Jupyter\7124274_netflix_logo_icon.png')
plt.figimage(logo, 0, 0, zorder=0, alpha=0)
imagebox = OffsetImage(logo, zoom= 0.4)
ab = AnnotationBbox(imagebox, (0.5, 1.2), frameon=False, xycoords='axes fraction', boxcoords="axes fraction", pad=0)
plt.gca().add_artist(ab)
text_box.set_bbox(dict(facecolor='red', alpha= 0.5, edgecolor = 'red'))
plt.xlabel('TOTAL CONTENT STREAMED', color = 'red')
plt.ylabel('COUNTRIES', color = 'red')
plt.show()
```

```python
df['Genres'] = df['Listed_In'].apply(lambda x: x.split(','))
top_genres = pd.Series(sum(df['Genres'], [])).value_counts().head(5)
top_genres
```

```python
#We need a slight change in here. The topmost director plot is shown by not given which is a bit inconvenient.
#We will change that.
df['Director'] = df['Director'].replace('Not Given' , 'Unknown director')
```

```python
#Lets run the value count again. Perfect!
directors = df['Director'].value_counts().head(10)
directors
```

```python
#Visualizing the top 10 most title-winning directors 
plt.figure(figsize= (8, 6))
sns.barplot(x= directors.values, y = directors.index, hue = directors, palette = 'bright')
plt.title('TOP 10 MOST TITLE-WINNING DIRECTORS [2008 - 2020]',  fontdict={'fontsize': 14, 'fontweight': 'bold', 'fontname': 'Arial', 'color': 'red'})
logo = mpimg.imread(r'C:\Users\hisar\OneDrive\Desktop\Jupyter\7124274_netflix_logo_icon.png')
plt.figimage(logo, 0, 0, zorder=0, alpha=0)
imagebox = OffsetImage(logo, zoom= 0.3)
ab = AnnotationBbox(imagebox, (0.5, 1.2), frameon=False, xycoords='axes fraction', boxcoords="axes fraction", pad=0)
plt.gca().add_artist(ab)
plt.xlabel('NUMBER OF TITLES WON', color = 'red')
plt.ylabel('NAME OF DIRECTORS',color = 'red')
plt.show()
```

```python
years = df['Year_Added'] = df['Date_Added'].dt.year #To seperate years from the given dates.
```

```python
year_addons = df['Year_Added'].value_counts().sort_index()
year_addons 
```

```python
#Visualizing the annual yearly trends in content add-ons
plt.figure(figsize =(8,6))
sns.lineplot(x = year_addons.index , y = year_addons.values, marker = 'o', color = 'blue')
text_box= plt.title('THE ANNUAL YEARLY TRENDS BASED ON CONTENT ADD-ON:', fontweight = 'bold', fontsize = 12, color = 'black' )
text_box.set_bbox(dict(facecolor='red', alpha= 1, edgecolor = 'red'))
logo = mpimg.imread(r'C:\Users\hisar\OneDrive\Desktop\Jupyter\pngwing.com.png')
plt.figimage(logo, 0, 0, zorder=0, alpha=0)
imagebox = OffsetImage(logo, zoom= 0.08)
ab = AnnotationBbox(imagebox, (0.5, 1.2), frameon=False, xycoords='axes fraction', boxcoords="axes fraction", pad=0)
plt.gca().add_artist(ab)
plt.xlabel('YEARS HAVING ADD-ONs', color= 'red')
plt.ylabel('FREQUENCY OF CONTENTS', color = 'red')
plt.grid(True)
plt.show()
           
```

```python
df['Month_Added'] = df['Date_Added'].dt.month
```

```python
month_addons = df['Month_Added'].value_counts().sort_index()
month_addons
```

```python
#Visualizing the monthly content add-ons 
plt.figure(figsize =(8,6))
sns.barplot(x= month_addons.index, y= month_addons.values, hue=month_addons, palette = 'pastel', legend=False)
text_box= plt.title('THE MONTHLY TRENDS BASED ON CONTENT ADD-ON:', fontweight = 'bold', fontsize = 12, color = 'black' )
text_box.set_bbox(dict(facecolor='red', alpha= 1, edgecolor = 'red'))
logo = mpimg.imread(r'C:\Users\hisar\OneDrive\Desktop\Jupyter\pngwing.com.png')
plt.figimage(logo, 0, 0, zorder=0, alpha=0)
imagebox = OffsetImage(logo, zoom= 0.08)
ab = AnnotationBbox(imagebox, (0.5, 1.2), frameon=False, xycoords='axes fraction', boxcoords="axes fraction", pad=0)
plt.gca().add_artist(ab)
plt.xlabel('MONTHS', color= 'red')
plt.ylabel('FREQUENCY OF CONTENT ADD-ONs', color = 'red')
plt.xticks(range(1, 13), ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'])
plt.show()
```
